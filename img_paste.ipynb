```python

# import PIL module 
from PIL import Image 
import sys

# Total number of frames: the background images and the front body images must have the same number of frames!
for i in range(0, 46):
    # get index 
    if len(str(i)) == 1:
        index = '0' + str(i)
    else:
        index = str(i)
        
    # Front Image
    filename = fr"fb\0000{index}.png"
    
    # Back Image 
    filename1 = fr"bg\frame_{index}_delay-0.14s.gif"
    filename2 = fr"bg\frame_{index}_delay-0.15s.gif"
    
    # Open Front Image, resize it to appropriate size
    base_width = 800
    frontImage = Image.open(filename) 
    wpercent = (base_width / float(frontImage.size[0]))
    hsize = int((float(frontImage.size[1]) * float(wpercent)))
    frontImage = frontImage.resize((base_width, hsize))
    
    # Open Background Image 
    try:
        background = Image.open(filename1) 
    except:
        background = Image.open(filename2) 
    
    # Convert image to RGBA 
    frontImage = frontImage.convert("RGBA") 
    
    # Convert image to RGBA 
    background = background.convert("RGBA") 
    
    # Adjust the front image position with respect to the background image
    
    # Calculate width
    width = (background.width - frontImage.width)//2 -20
    
    # Calculate height 
    height = (background.height - frontImage.height)
    
    # Paste the frontImage at (width, height) 
    background.paste(frontImage, (width, height), frontImage) 

    # Resize the image to 630x820
    background = background.resize((630, 820))

    print("background.size", background.size)
    # Save this image 
    background.save("new.png",quality=1, optimize=True)

    new_image = Image.open("new.png")
    img_left =  new_image.crop((0, 0, 506, 820))
    img_right = new_image.crop((530, 0, 630, 820))
    print("img_left.size", img_left.size)
    print("img_right.size", img_right.size)
    img_left.save(f"img_left{index}.png")
    img_right.save(f"img_right{index}.png")
    # For test, break the loop
    break
```
